# HW05 — Data Storage (CSV + Parquet with env-driven paths)


```python
import os, sys
from pathlib import Path
import pandas as pd

sys.path.append("src")
from config import get_data_dir_raw, get_data_dir_processed
from storage import write_df, read_df, build_path_raw, build_path_processed

RAW = get_data_dir_raw(); RAW.mkdir(parents=True, exist_ok=True)
PRO = get_data_dir_processed(); PRO.mkdir(parents=True, exist_ok=True)

# Sample DataFrame
df = pd.DataFrame({"a":[1,2,3], "b":[0.1, 0.2, 0.3]})

csv_path = build_path_raw("sample.csv")
pq_path  = build_path_processed("sample.parquet")

write_df(df, csv_path)
write_df(df, pq_path)

df_csv = read_df(csv_path)
df_pq  = read_df(pq_path)

print("Shapes:", df_csv.shape, df_pq.shape)
print("Dtypes CSV:", df_csv.dtypes.to_dict())
print("Dtypes PQ:", df_pq.dtypes.to_dict())

# Simple validation
assert df_csv.shape == df_pq.shape
assert set(df_csv.columns) == set(df_pq.columns)
print("Validation OK")
```
