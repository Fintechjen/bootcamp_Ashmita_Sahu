# HW03 — Python Fundamentals


Covers NumPy, pandas loading/inspection, summary stats, groupby, and saving outputs.


```python
import os
import pandas as pd
import numpy as np
from pathlib import Path

# Try to load provided CSV, else create a small sample.
path = Path("data/starter_data.csv")
if not path.exists():
    Path("data").mkdir(exist_ok=True)
    df = pd.DataFrame({
        "category": ["A","A","B","B","C"],
        "value": [10, 12, 7, 8, 15],
        "score": [0.1, 0.2, 0.3, 0.2, 0.5],
    })
    df.to_csv(path, index=False)
else:
    df = pd.read_csv(path)

df.info()
display(df.head())

# Summary stats
summary = df.describe(include='number').T
display(summary)

# Groupby
gb = df.groupby("category", as_index=False).agg(
    value_mean=("value","mean"),
    score_sum=("score","sum")
)
display(gb)

# Save outputs
Path("data/processed").mkdir(parents=True, exist_ok=True)
summary.to_csv("data/processed/summary.csv", index=False)
print("Saved to data/processed/summary.csv")
```
