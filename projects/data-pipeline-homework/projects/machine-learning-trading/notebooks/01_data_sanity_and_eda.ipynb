# 01 — Data sanity + quick EDA


```python
import pandas as pd, numpy as np
import yfinance as yf
import matplotlib.pyplot as plt

ticker = "AAPL"
df = yf.download(ticker, period="1y", interval="1d", progress=False).reset_index()
df.info()
display(df.head())

# Simple plot
df.set_index("Date")["Close"].plot(title=f"{ticker} Close (1y)")
plt.show()
```
